```python
communities = [40, 80, 60]
G = graphs.Community(N=180, Nc=3, comm_sizes=communities)
f = np.random.normal(size=G.N)
G.compute_fourier_basis()

f_hat = G.gft(f)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
G.plot_signal(f, vertex_size=30, ax=axes[0])
axes[1].step(G.e, np.abs(f_hat), '.-')
plt.show()
```

```python
fp = np.zeros(G.N)
fp[:communities[0]] = -1 * np.ones(communities[0])
fp[-communities[-1]:] = 1 * np.ones(communities[-1])

fp_hat = G.gft(fp)
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
G.plot_signal(f, vertex_size=30, ax=axes[0])
G.plot_signal(fp, vertex_size=30, ax=axes[1])

axes[2].step(G.e, np.abs(f_hat), '.-b')
axes[2].step(G.e, np.abs(fp_hat), '.-r')
plt.show()
```

```python
"""Make a graph to de-noise using the De-noise optimization"""
# Generate a graph
G = graphs.Sensor()
# compute the laplacian
G.compute_laplacian()

G.compute_fourier_basis()
# make the signal to be placed on the graph
f_sig = G.U[0] + G.U[1] + G.U[2]
f_sig_noise = f_sig + 1 * np.random.randn(f_sig.shape[0])

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

G.plot_signal(f_sig_noise, vertex_size=30, ax=axes[0])

# denoise 
def denoise_opt(y, lapl, alpha):
    
    inter_mat = np.eye(y.shape[1]) + alpha * lapl
    
    x = np.linalg.solve(inter_mat, y.T)    
    return x

alf = 20

x_hat = denoise_opt(f_sig_noise.reshape(1, -1), G.L, alpha=alf)

def rms_error(y, x):
        
        error_vec = y - x
        rms_e = np.sqrt(np.mean(np.square(error_vec)))
        
        return rms_e

def cost_denoise(y, x, alpha, lapl):
    
    diff_vec = y.T - x.T
    diff_vec = diff_vec.T
    
    print(diff_vec.shape)
    
    cost = diff_vec.T @ diff_vec + alpha*x.T @ lapl @ x
    return cost

G.plot_signal(x_hat, vertex_size=30, ax=axes[1])
plt.show()

plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(f_sig_noise, 'b')
plt.plot(f_sig, 'k')
plt.plot(x_hat, 'r')

plt.legend(('noisy signal', 'original smooth signal', 'de-noise signal'))

plt.show()

rms_ee = rms_error(f_sig, x_hat)

cost_val = cost_denoise(f_sig.T, x_hat, alf, G.L)


print(rms_ee, cost_val)
```

# Assignment: Inpainting

```python
"""Inpaint a sensor"""

# Generate a graph
G = graphs.Sensor()
# compute the laplacian
G.compute_laplacian()

G.compute_fourier_basis()
# make the signal to be placed on the graph
f_sig = G.U[0] + G.U[1] + G.U[2]
# normalize both the original signal
f_sig = f_sig.reshape(-1, 1) / np.amax(np.absolute(f_sig))

# add gaussian noise if req'd
f_sig_noise = f_sig + 1 * np.random.randn(f_sig.shape[0])

# make the vector for the removing elements
p = 0.8
N = f_sig.shape[0]

zero_ones = [0]*int(p*N) + [1]*int(N-N*p)

arr_removal = np.random.permutation(zero_ones)

if len(arr_removal) != f_sig.shape[0]:
    
    diff_len = f_sig.shape[0] - len(arr_removal)
    arr_removal = np.append(arr_removal, [1]*diff_len)
    

# make M matrix : diag(1 or 0)
m_mat = np.diag(arr_removal)

# salt and pepper noise: missing channels
f_s_p_noise = f_sig.copy()
f_s_p_noise[np.where(arr_removal==0)] = 0


plt.step(np.linspace(0, 1, f_s_p_noise.shape[0]), f_s_p_noise)
plt.step(np.linspace(0, 1, f_s_p_noise.shape[0]), f_sig)
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
G.plot_signal(f_sig, vertex_size=30, ax=axes[0])
G.plot_signal(f_s_p_noise, vertex_size=30, ax=axes[1])

plt.show()
```

```python
def inpaint_optima(y, mmat, alfa, lapl):
    
    # sp_mmat = sp.sparse.csr_matrix(mmat)
    inter_mat = (mmat + alfa*lapl)
    
    y_inter_mat = y
    
    x = np.linalg.solve(inter_mat, y_inter_mat)
    return x


def build_m_mat(flatted_img):
    diag_arr = list(0 if y == 0 else 1 for y in flatted_img)
    m_mat = np.diag(diag_arr)
    return m_mat


alpha = 0.01
inpaint_x = inpaint_optima(f_s_p_noise, m_mat, alpha, G.L)

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    
plt.step(np.linspace(0, 1, f_s_p_noise.shape[0]), f_s_p_noise, 'k')
plt.step(np.linspace(0, 1, f_s_p_noise.shape[0]), f_sig, 'b')
plt.step(np.linspace(0, 1, f_s_p_noise.shape[0]), inpaint_x, 'r')
plt.legend(('lossy signal', 'original signal', 'inpainted signal'))

plt.savefig('restored_sig_large_alpha.pdf')

plt.show()
```

```python

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
G.plot_signal(inpaint_x, vertex_size=30, ax=axes[0])
G.plot_signal(f_sig, vertex_size=30, ax=axes[1])
G.plot_signal(f_s_p_noise, vertex_size=30, ax=axes[2])

plt.savefig('graphs.pdf')

plt.show()
```

# Improve Resolution of a Low Res image using Inpainting

```python
rsz_tup = (64, 64)

img_graph = graphs.Grid2d(N1=rsz_tup[0], N2=rsz_tup[1])
f = sp.misc.face(gray=True)

f_low_res = sp.misc.imresize(f, rsz_tup)

plt.imshow(f_low_res, cmap=plt.cm.gray)
plt.savefig('low_res_raccoon.pdf')

img_graph.plot_signal(f_low_res.flatten())


plt.show()
```

```python
def internal_pad(aray, n):
    x, y = aray.shape
    indices = np.repeat(np.arange(y+1), n)[n:-n]
    stretch_x = np.insert(aray, indices, 0, axis=1)
    stretch_y = np.insert(stretch_x, indices, 0, axis=0)
    return stretch_y


expand_img = internal_pad(f_low_res, 1)

flat_img = expand_img.flatten().reshape(-1, 1)

plt.imshow(expand_img, cmap=plt.cm.gray)
plt.savefig('internal_padding.pdf')
plt.show()
```

```python
shape_img = expand_img.shape

im_rest_graph = graphs.Grid2d(N1=shape_img[0], N2=shape_img[1])
im_rest_graph.compute_laplacian()

im_rest_graph.plot_signal(expand_img.flatten())

plt.show()
```

```python
imgm_mat = build_m_mat(flat_img)
alfa = 0.001

restore_flat = inpaint_optima(flat_img, imgm_mat, alfa, im_rest_graph.L)

restore_img = restore_flat.reshape(expand_img.shape[0], expand_img.shape[1])
```

```python
plt.imshow(restore_img, cmap=plt.cm.gray)
plt.show()
```

```python
imgm_mat = build_m_mat(flat_img)
alfa = 1

restore_flat = inpaint_optima(flat_img, imgm_mat, alfa, im_rest_graph.L)

restore_img = restore_flat.reshape(expand_img.shape[0], expand_img.shape[1])

plt.imshow(restore_img, cmap=plt.cm.gray)
plt.show()
```
